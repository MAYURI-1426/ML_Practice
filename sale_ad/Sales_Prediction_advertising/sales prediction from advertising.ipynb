```python
import pandas as pd 
import seaborn as sns 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("/content/advertising.csv")
data.head()
```

```python
data.isna().sum()
```

```python
sns.heatmap(data.corr(), square=True)
```

```python
sns.lmplot(x='TV', y='Sales', data=data)
```

```python
X = data[['TV']]
y = data.Sales
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.4, random_state=20
)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
model.predict([[20]])
```

```python
y_prediction = model.predict(X_test)
```

```python
act_prediction= pd.DataFrame({
    'Actual_val': y_test.values.flatten(), 
    'Predict_val': y_prediction.flatten()
})

act_prediction.head(5)
```

```python
act_prediction.sample(10).plot(kind='bar')
```

```python
model.score(X_test, y_test)
```
