```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

ipl_data = pd.read_csv("/content/Book_ipl22_ver_33.csv")
print(ipl_data.head())
```

```python
#number of matches won by each team in IPL 2022:

figure = px.bar(ipl_data, x=ipl_data["match_winner"],
            title="Number of Matches Won in IPL 2022")
figure.show()
```

```python
#whether most of the teams win by defending (batting first) or chasing 

ipl_data["won_by_1"] = ipl_data["won_by"].map({"Wickets": "Chasing", 
                                     "Runs": "Defending"})
won_by1 = ipl_data["won_by_1"].value_counts()
label = won_by1.index
counts = won_by1.values
colors = ['gold','blue']

fig = go.Figure(data=[go.Pie(labels=label, values=counts)])
fig.update_layout(title_text='Number of Matches Won By Defending Or Chasing')
fig.update_traces(hoverinfo='label+percent', textinfo='value', 
                  textfont_size=30,
                  marker=dict(colors=colors, 
                              line=dict(color='black', width=4)))
fig.show()
```

```python
#what most teams prefer (batting or fielding) after winning the toss:

toss = ipl_data["toss_decision"].value_counts()
label = toss.index
counts = toss.values
colors = ['skyblue','yellow']

fig = go.Figure(data=[go.Pie(labels=label, values=counts)])
fig.update_layout(title_text='Toss Decision')
fig.update_traces(hoverinfo='label+percent', 
                  textinfo='value', textfont_size=30,
                  marker=dict(colors=colors, 
                              line=dict(color='black', width=3)))
fig.show()
```

```python
#the top scorers of most IPL 2022 matches:

figure = px.bar(ipl_data, x=ipl_data["top_scorer"],
            title="Top Scorers in IPL 2022")
figure.show()
```

```python
figure = px.bar(ipl_data, x=ipl_data["top_scorer"], 
                y = ipl_data["highscore"], 
                color = ipl_data["highscore"],
            title="Top Scorers in IPL 2022")
figure.show()
```

```python
#most player of the match awards till now in IPL 2022:

figure = px.bar(ipl_data, x = ipl_data["player_of_the_match"], 
                title="Most Player of the Match Awards")
figure.show()
```

```python
#bowlers with the best bowling figures in most of the matches:

figure = px.bar(ipl_data, x=ipl_data["best_bowling"],
            title="Best Bowlers in IPL 2022")
figure.show()
```

```python
figure = go.Figure()
figure.add_trace(go.Bar(
    x=ipl_data["venue"],
    y=ipl_data["first_ings_wkts"],
    name='First Innings Wickets',
    marker_color='gold'
))
figure.add_trace(go.Bar(
    x=ipl_data["venue"],
    y=ipl_data["second_ings_wkts"],
    name='Second Innings Wickets',
    marker_color='lightgreen'
))
figure.update_layout(barmode='group', xaxis_tickangle=-45)
figure.show()
```

```python
venue = ipl_data["venue"].value_counts()
label = venue.index
counts = venue.values
colors = ['orange','pink']

fig = go.Figure(data=[go.Pie(labels=label, values=counts)])
fig.update_layout(title_text='No. of matches')
fig.update_traces(hoverinfo='label+percent', 
                  textinfo='value', textfont_size=30,
                  marker=dict(colors=colors, 
                              line=dict(color='black', width=3))
)
fig.show()
```
